## Objectives
* Get a table of German Tatoeba Sentences that have audio associated with them.

```python
import pandas as pd
```

```python
deu_sentences = pd.read_csv('./deu_sentences.tsv', sep='\t', names=['sentence_id',
                                                   'lang',
                                                   'text'])
```

```python
audio_sentences = pd.read_csv('./sentences_with_audio.csv', sep='\t',
                              names=['sentence_id', 'audio_id', 'username',
                                     'licence', 'attribution_url'])
```

```python
audio_sentences
```

```python
deu_audio = pd.merge(audio_sentences, deu_sentences, on='sentence_id')
```

```python
# This table looks right. Searching for all German sentences
# with audio on the Tatoeba website returns 32 939 results
# on 1 Feb 2025. My data is a little older so the extra sentences
# are probably from new additions to the database.
deu_audio
```

```python
deu_audio.to_csv('./deu_audio.tsv', sep='\t', index=False)
```
